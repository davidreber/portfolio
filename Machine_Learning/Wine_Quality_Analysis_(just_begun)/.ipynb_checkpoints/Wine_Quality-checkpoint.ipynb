Thus, there do not appear to be any obvious correlations between "quality" and any of the other features. This observation is validated by a quick correlation check between all pairings of features. 

```python
cols = list(red) # column names are same for both datasets
correlation_threshold = .5

def correlations(df, name, threshold):
    """Prints correlations between features of df which are above threshold.
        Accepts: 
            df (DataFrame):      the dataframe to be examined
            name (str):          printed name of dataframe
            threshold (float):   threshold for correleation coefficient
    """
    corr = red.corr()
    corr *= np.triu(np.ones_like(corr),1) # ignore symmetric values
    masked = corr[np.abs(corr)>threshold]
    abv_thresh = list(zip(*np.where(masked.notnull())))
    print('\033[1m' + name + '\033[0m')
    print('{:>25}  {:>25}  {:>25}'.format("Feature","Feature","Correlation Coefficient"))
    for tup in abv_thresh:
        i,j = tup
        print('{:>25}  {:>25}  {:>25}'.format(cols[i],cols[j],masked.values[i,j]))

# Print Red Wine features with correlation coeficient above/below theshold
correlations(red, "Red Wine Correlations", .5)

# Print White Wine features with correlation coeficient above/below theshold
correlations(white, "White Wine Correlations", .5)
```

Every correlation with a correlation coeficient above/below 0.5 or -0.5 is an obvious one, for example the strong correlations between acid concentrations and pH. Note that there is no predictive feature for "quality" with correlation above 0.5.

## Outlier Discussion

The discrepency between the mean and maximum values of features such as "residual sugar" and "free sulfur dioxide" indicates that there are outliers in our datasets.



iterate through each row. If an entry in the row is more than 3 stds from mean of col, flag row as an outlier.
Then, for each such row, print the row number, the entry's col name, and the entry value. (If multiple such outliers in row, only report first one)

```python
def remove_outliers(df):
    "Removes the outliers of df, as marked by having a feature value more than 3 std. deviations from mean."
    new_df = df.copy()
    cols = list(df)
    n = np.shape(df)[0]
    for j, col in enumerate(cols):
        new_df[np.abs(df[col]-df[col].mean())>(3*df[col].std())] = np.nan
    new_df = new_df.dropna()
    return new_df

new_red = remove_outliers(red)
print(np.shape(red)[0] - np.shape(new_red)[0])
```

However, note that all the features (except "quality") are based solely on measurements the contents of the wine. Especially since this dataset was systematically produced, there is no reason to believe outside bias affected the measurement of these primary features (as recent weather might affect harvest yields, etc). Thus, rather than removing the outliers as erronous data, we will retain them, but focus on adding appropriate regulariztion to our regression.

## Prediction of Wine Quality

Having performed initial analysis on the dataset, we now prepare to predict the quality of the wines. To begin with, we split our data into output features ("quality"), and input features (everything else).

```python
red_X = red[cols[:-1]]
red_y = red[cols[-1]]
white_X = white[cols[:-1]]
white_y = white[cols[-1]]
```

The following function will help us avoid repetitious code. The function accepts the input and output data, and over a number of trials performs a test-training split on the data, and runs several predictive models.

While we focus on linear regression (with L-1 and L-2 regularization), we also use a Random Forest for comparison of model complexity. Since the output variable of "quality" is comparative, we focus on regression techniques rather than classification.

```python
import sklearn.linear_model as lm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
    
def learn(X, y, method, num_runs, tol, train_split=0.7):
    accs = []
    weights = []
    for i in range(num_runs):
        train_x, test_x, train_y, test_y = train_test_split(X, y,train_size=train_split)
        method.fit(train_x, train_y)
        accs.append(np.mean(np.abs(method.predict(test_x)-test_y) < tol))
    return np.mean(accs), np.std(accs)

methods = [lm.LinearRegression(), lm.Lasso, lm.Ridge(), RandomForestRegressor()]
num_runs = [100,100,100,50]
names = ["Linear Regression","Lasso Regression", "Ridge Regression","Random Forest Regression"]
```

We begin by running our regression with an acceptable tolerance for error of 0.5, which corresponds to guessing and rounding up or down to the correct answer:

```python
tol = 0.5

print('{:>25}  {:>25}  {:>25}'.format("Method","Accuracy"," Standard Dev. (of Acc.)"))
for i, method in enumerate(methods):
    mean_acc, std_acc = learn(red_X,red_y,lm.LinearRegression(),num_runs[i],tol)
    print('{:>25}  {:>25}  {:>25}'.format(names[i],mean_acc,std_acc))
```

Allowing for a larger tolerance of 1, obtain 88% accuracy. Note that the Random Forest doesn't work much better on this problem, which fits the complexity of the problem.

```python
tol = 1.0

print('{:>25}  {:>25}  {:>25}'.format("Method","Accuracy"," Standard Dev. (of Acc.)"))
for i, method in enumerate(methods):
    mean_acc, std_acc = learn(red_X,red_y,lm.LinearRegression(),num_runs[i],tol)
    print('{:>25}  {:>25}  {:>25}'.format(names[i],mean_acc,std_acc))
```

However, based on our previous visualizations, there seem to be some linear dependence between the features. To remove this dependence, we perform a transform using PCA, and retrain our models:

```python
from sklearn.decomposition import PCA

Zred_X = np.zeros_like(red_X)
X = np.array(red_X)
for j in range(len(X.T)):
    Zred_X[:,j] = (X[:,j]-np.mean(X[:,j]))/np.std(X[:,j])
    
pca = PCA()
pca.fit(Zred_X)
new_red_X = pca.transform(Zred_X)
```

```python
for i, method in enumerate(methods):
    mean_acc, std_acc = learn(new_red_X,red_y,lm.LinearRegression(),num_runs[i],tol)
    print('{:>25}  {:>25}  {:>25}'.format(names[i],mean_acc,std_acc))
```

This transformation improved Ridge and Random Forest regressions, but worsened the Linear and Lasso regressions.
